```python
from sklearn.datasets import make_moons
from sklearn.model_selection import RandomizedSearchCV
import matplotlib.pyplot as plt
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

Let's generate our dataset and take a look on it.

```python
moons_points, moons_labels = make_moons(n_samples=500, noise=0.2, random_state=42)
plt.scatter(moons_points[:, 0], moons_points[:, 1], c=moons_labels)
```

## 1.1 Pure models.
First let's try to solve this case with good old Logistic Regression and simple (linear kernel) SVM classifier.

Train LR and SVM classifiers (choose params by hand, no CV or intensive grid search neeeded) and plot their decision regions. Calculate one preffered classification metric.

Describe results in one-two sentences.

_Tip:_ to plot classifiers decisions you colud use either sklearn examples ([this](https://scikit-learn.org/stable/auto_examples/neural_networks/plot_mlp_alpha.html#sphx-glr-auto-examples-neural-networks-plot-mlp-alpha-py) or any other) and mess with matplotlib yourself or great [mlxtend](https://github.com/rasbt/mlxtend) package (see their examples for details)

_Pro Tip:_ wirte function `plot_decisions` taking a dataset and an estimator and plotting the results cause you want to use it several times below

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC

from mlxtend.plotting import plot_decision_regions

from sklearn.metrics import f1_score

lr = LogisticRegression(solver='lbfgs', C=10)# add some params
svm = SVC(kernel='linear', C=100) # here too

lr.fit(moons_points, moons_labels)
svm.fit(moons_points, moons_labels)
print('F1 score for logistic regression: {}'.format(f1_score(lr.predict(moons_points), moons_labels)))
print('F1 score for SVM: {}'.format(f1_score(svm.predict(moons_points), moons_labels)))

plt.figure(figsize=(16, 6))
plt.subplot(121)
plot_decision_regions(moons_points, moons_labels, clf=lr, legend=2)
plt.title("Decision regions for LogisticRegression")
plt.subplot(122)
plot_decision_regions(moons_points, moons_labels, clf=svm, legend=2)
plt.title("Decision regions for SVM")
plt.show()
```

### Description

We see that both LogisticRegression and with simple SVM classifiers built linear dividing surface, which we expected. Also we should note, that we get quite good F1 score for both algorithms, but they do not represent actual distribution.

## 1.2 Kernel trick

Now use different kernels (`poly`, `rbf`, `sigmoid`) on SVC to get better results. Play `degree` parameter and others.

For each kernel estimate optimal params, plot decision regions, calculate metric you've chosen eariler.

Write couple of sentences on:

* What have happenned with classification quality?
* How did decision border changed for each kernel?
* What `degree` have you chosen and why?

Let's create functions that will conduct experiments with SVM with different kernels and parameters. Then we will have a look at decision plots and analyze them.

```python
def conduct_experiment_degree(kernel, params, data, labels):
    metric_scores = []
    plt.figure(figsize=(18, 10))
    
    for i in range(len(params)):
        svm = SVC(kernel=kernel, degree=params[i])
        svm.fit(data, labels)
        metric_scores.append(f1_score(svm.predict(data), labels, average='weighted'))
        plt.subplot(2, 3, i+1)
        plot_decision_regions(data, labels, clf=svm, legend=2)
        plt.title("Decision regions for SVM with {} kernel, deg = {}".format(kernel, params[i]))
    
    plt.show()
    return metric_scores
```

```python
def conduct_experiment_C(kernel, params, data, labels):
    metric_scores = []
    plt.figure(figsize=(18, 10))
    
    for i in range(len(params)):
        svm = SVC(kernel=kernel, C=params[i])
        svm.fit(data, labels)
        metric_scores.append(f1_score(svm.predict(data), labels, average='weighted'))
        plt.subplot(2, 3, i+1)
        plot_decision_regions(data, labels, clf=svm, legend=2)
        plt.title("Decision regions for SVM with {} kernel, C = {}".format(kernel, round(params[i], 2)))
    
    plt.show()
    return metric_scores
```

```python
def conduct_experiment_gamma(kernel, params, data, labels):
    metric_scores = []
    plt.figure(figsize=(18, 10))
    
    for i in range(len(params)):
        svm = SVC(kernel=kernel, gamma=params[i])
        svm.fit(data, labels)
        metric_scores.append(f1_score(svm.predict(data), labels, average='weighted'))
        plt.subplot(2, 3, i+1)
        plot_decision_regions(data, labels, clf=svm, legend=2)
        plt.title(r"Decision regions for SVM with {} kernel, $\gamma$ = {}".format(kernel, round(params[i], 2)))
    
    plt.show()
    return metric_scores
```

```python
# differrent parameters for SVM
degrees = np.arange(0, 6, 1)
array_C = np.logspace(-1, 1, 6)
gammas = np.logspace(-1, 1, 6)
```

### Experiments with poly kernel

```python
metric_scores = conduct_experiment_degree('poly', degrees, moons_points, moons_labels)
print('F1 scores for different degrees with poly kernel: \n{}'.format(np.around(np.array(metric_scores), 3)))
```

```python
metric_scores = conduct_experiment_C('poly', array_C, moons_points, moons_labels)
print('F1 scores for different regularization parameters C with poly kernel: \n{}'.format(np.around(np.array(metric_scores), 3)))
```

```python
metric_scores = conduct_experiment_gamma('poly', gammas, moons_points, moons_labels)
print('F1 scores for different gamma parameters with poly kernel: \n{}'.format(np.around(np.array(metric_scores), 3)))
```

### Experiments with rbf kernel

```python
metric_scores = conduct_experiment_C('rbf', array_C, moons_points, moons_labels)
print('F1 scores for different regularization parameters C with rbf kernel: \n{}'.format(np.around(np.array(metric_scores), 3)))
```

```python
metric_scores = conduct_experiment_gamma('rbf', gammas, moons_points, moons_labels)
print('F1 scores for different gamma parameters with rbf kernel: \n{}'.format(np.around(np.array(metric_scores), 3)))
```

### Experiments with sigmoid kernel

```python
metric_scores = conduct_experiment_C('sigmoid', array_C, moons_points, moons_labels)
print('F1 scores for different regularization parameters C with sigmoid kernel: \n{}'.format(np.around(np.array(metric_scores), 3)))
```

```python
metric_scores = conduct_experiment_gamma('sigmoid', gammas, moons_points, moons_labels)
print('F1 scores for different gamma parameters with sigmoid kernel: \n{}'.format(np.around(np.array(metric_scores), 3)))
```

### Answers

* What have happenned with classification quality? It depends on kernel and tuned parameter. For example, for `poly` kernel the best metrics showed classifier with degree of 1 and 3, and the quality didn't increase when we incremented the degree. However, for `rbf` kernel metric increased while increasing C parameter, which can be considered as overfitting due to plots.
* How did decision border changed for each kernel? For `poly` kernel decision border changed due to the degree of polynom while changing the `degree` parameters. In some cases it changed to fit data more properly (in cases of `C` and `gamma` parameters).
* What `degree` have you chosen and why? I would choose the `degree` to be equal to 1 or 3 in `poly` kernel, because it seems that decision borders are more close to expected and these parameters showed the best metrics result.

## 1.3 Simpler solution (of a kind)
What is we could use Logisitc Regression to successfully solve this task?

Feature generation is a thing to help here. Different techniques of feature generation are used in real life, couple of them will be covered in additional lectures.

In particular case simple `PolynomialFeatures` ([link](https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.PolynomialFeatures.html)) are able to save the day.

Generate the set of new features, train LR on it, plot decision regions, calculate metric.

* Compare SVM's results with this solution (quality, borders type)
* What degree of PolynomialFeatures have you used? Compare with same SVM kernel parameter.

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import Pipeline
```

```python
def conduct_experiment_poly_features(data, labels, **kwargs):
    degrees = np.arange(2, 10, 1)
    metric_scores = []
    plt.figure(figsize=(18, 30))
    
    for i in range(len(degrees)):
        log_reg_pipeline = Pipeline(steps=(('preproc', PolynomialFeatures(degrees[i])), 
                               ('estimator', LogisticRegression(**kwargs))))

        log_reg_pipeline.fit(data, labels)
        plt.subplot(5, 2, i+1)
        plot_decision_regions(data, labels, clf=log_reg_pipeline, legend=2)
        plt.title("Decision regions for LogReg with PolyFeature deg = {}".format(degrees[i]))
        metric_scores.append((f1_score(log_reg_pipeline.predict(data), labels, average='weighted')))
        
    plt.show()  
    return metric_scores
```

```python
metric_scores = conduct_experiment_poly_features(moons_points, moons_labels, solver='lbfgs', C=10)
print('F1 scores for different degrees of polynomial features in logistic regression: \n{}'.format(np.around(np.array(metric_scores), 3)))
```

### Answers

* Compare SVM's results with this solution (quality, borders type). Logistic regression with polynomial features and SVM with `rbf` kernel obtained almost equally good results, but SVM with other kernels demostrated worse performance.
* What degree of PolynomialFeatures have you used? Compare with same SVM kernel parameter. It looks like odd degrees of polynomial features were the best due to metric ans decision regions.
